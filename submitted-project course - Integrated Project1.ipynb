## Step 5. Test the following hypotheses:

### Average user ratings of the Xbox One and PC platforms are the same. null hypothesis

<div style="border:solid orange 2px; padding: 5px"; class="alert alert-info">   
<span style="color:brown; font-size: 15px"> <b>
It is a form that assumes no relationship between the variables. In other words, assume a treatment has no effect.

```python
Xbox=store_Ice1[store_Ice1.platform=='XOne']
PC=store_Ice1[store_Ice1.platform=='PC']
```

```python
st.levene(Xbox.sum_sale,PC.sum_sale,center='mean') # very low pvalue
```

```python
alpha = 0.05  # critical statistical significance

result1 = st.ttest_ind(Xbox.sum_sale,PC.sum_sale, equal_var=False)
print('p-value: ', result1.pvalue)

if result1.pvalue < alpha:
    print('We reject the null hypothesis')
else:
    print("We can't reject the null hypothesis") 
# general mean
Xbox_mean=Xbox.sum_sale.mean()
PC_mean=PC.sum_sale.mean()
print('Xbox mean',Xbox_mean)
print('PC mean',PC_mean)
```

### Average user ratings for the Action and Sports genres are different.

<div style="border:solid orange 2px; padding: 5px"; class="alert alert-info">   
<span style="color:brown; font-size: 15px"> <b>
hypothesis test is to demonstrate that in the given condition, there is sufficient evidence supporting the credibility of alternative hypothesis instead of null hypothesis test

```python
Action=store_Ice1[store_Ice1.genre=='Action']
Sports=store_Ice1[store_Ice1.genre=='Sports']
```

```python
st.levene(Action.sum_sale,Sports.sum_sale,center='mean') # high pvalue
```

```python
alpha = 0.05  # critical statistical significance

result1 = st.ttest_ind(Action.sum_sale,Sports.sum_sale, equal_var=True)
print('p-value: ', result1.pvalue)

if result1.pvalue < alpha:
    print('We reject the null hypothesis')
else:
    print("We can't reject the null hypothesis") 
# general mean
Action_mean=Action.sum_sale.mean()
Sports_mean=Sports.sum_sale.mean()
print('Action mean',Action_mean)
print('Sports mean',Sports_mean)
```
